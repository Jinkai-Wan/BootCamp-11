```python
df = pd.read_csv("california_housing.tsv", sep="\t")
# cause our file is .tsv, so we add sep="\t"
```

```python
df
```

```python
df = pd.read_csv("california_housing.tsv", sep="\t", index_col=0)
# cause our file is .tsv, so we add sep="\t"
# cause we know which index, we add index_col=0
```

```python
df
```

```python

california_ols = sm.OLS(df["MedHouseVal"],df.loc[:,["MedInc"]])
```

```python
california_ols_result = california_ols.fit()
```

```python
print(california_ols_result.summary())
```

```python

california_ols = sm.OLS(df["MedHouseVal"],sm.add_constant(df.loc[:,["MedInc","Latitude"]]))
```

```python
california_ols_result = california_ols.fit()
print(california_ols_result.summary())
```
